```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the first training session's results
session1_path = "/content/drive/MyDrive/Colab Notebooks/Luan Van/plotting/results.csv"
session2_path = "/content/drive/MyDrive/Colab Notebooks/Luan Van/plotting/results2.csv"

# Assuming the results are stored in CSV format
session1_data = pd.read_csv(session1_path)
session2_data = pd.read_csv(session2_path)

# Get the common columns (excluding 'epoch')
common_columns = list(set(session1_data.columns) & set(session2_data.columns) - {'epoch'})

# Create a figure and axes
fig, axes = plt.subplots(len(common_columns), 1, figsize=(10, 6 * len(common_columns)), sharex=True)  # Share x-axis

# Iterate through each common column and plot
for i, column in enumerate(common_columns):
    ax = axes[i]  # Get the current subplot axes
    ax.plot(session1_data['epoch'], session1_data[column], label=f"Session 1 - {column}")
    ax.plot(session2_data['epoch'], session2_data[column], label=f"Session 2 - {column}", linestyle='--')

    ax.set_title(f"{column} Comparison")
    ax.set_ylabel(column)
    ax.legend()
    ax.grid(True)

# Set x-axis label for the bottom subplot
axes[-1].set_xlabel("Epoch")

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig("all_metrics_comparison.png")  # Save the figure to a file
plt.show()
```
